```python
global book
book=[]
    
def train_search_book():
    time = float(input('원하는 시간 : 예) 6.15 (뒤로가기를 희망하면 99를 입력해주세요.) \n '))
    
    if time == 99:
        return
    start_station,end_station,train_type=input('출발역, 도착역, 열차종류를 입력하세요.\n').split(' ')
   
    f = open('TrainList.txt','r',encoding='utf-8')
    trainlist = f.readlines()
    for i in range(0, len(trainlist)) :
        trainlist[i] = trainlist[i].strip().split(' ') 
    edited_lines1 = []
    
    for i in range(1,len(trainlist)):
        if start_station==trainlist[i][1]:
            edited_lines1.append(trainlist[i])
    edited_lines2 = []
    
    for i in range(0,len(edited_lines1)):
        if end_station==edited_lines1[i][3]:
            edited_lines2.append(edited_lines1[i])
    edited_lines3 = []
    
    for i in range(0,len(edited_lines2)):
        if train_type==edited_lines2[i][4]:
            edited_lines3.append(edited_lines2[i])
            
    edited_lines4=[]
    edited_lines5=[]
    new_line=edited_lines3
    
    for i in range(0,len(edited_lines3)):
        new_line[i][0]=edited_lines3[i][0].lstrip("0").replace(":",".")

        if float(new_line[i][0])-time >=0:
            edited_lines4.append(new_line[i][0])
        else:
            edited_lines5.append(new_line[i][0])
    
    if len(edited_lines4)==0:
        edited_lines5.sort()
        keynumber1=float(edited_lines5[0])
        for i in range(0,len(edited_lines3)):
            if keynumber1 == float(new_line[i][0]):
                edited_lines6=edited_lines3[i]
    else:
        edited_lines4.sort()
        keynumber2=float(edited_lines4[0])
        for i in range(0,len(edited_lines3)):
            if keynumber2 == float(new_line[i][0]):
                edited_lines6=edited_lines3[i]
    
    for i in range(0,len(new_line)):
        new_line[i][0]='0'+new_line[i][0]
        new_line[i][0]= new_line[i][0].replace(".",":")
        
    print(edited_lines6)
    print(trainlist)
    
    
    yesno = int(input('예매하시겠습니까? \n네):1 \n아니오):2 \n ')) 
    if yesno==1:
        f = open("Trainlist.txt",'w')
        for line in trainlist:
            edited_lines6[5]=str(edited_lines6[5])
            if edited_lines6 == line:
                edited_lines6[5]=int(edited_lines6[5])-1
                if edited_lines6[5]<0:
                    print('예약가능 좌석이 없습니다.')
                    edited_lines6[5]=int(edited_lines6[5])+1
                else:
                    book.append(edited_lines6)
                    
                print(edited_lines6)
                f.write(' '.join(map(str,edited_lines6)))
                f.write('\n')
            else:
                f.writelines(' '.join(line))
                f.write('\n')
        
def train_book():
    check = int(input('예매현황확인:1 예매취소:2 (뒤로가기를 희망하면 99를 입력해주세요.)\n')) 
    if check==1:
        for i in range(0,len(book)):
            print(book)
    elif check==2:
        for i in range(0,len(book)):
            print(book)
            search = int(input('취소하고자하는 리스트 번호를 입력하시오. 예)[[1],[2],[3]](취소종료시 99입력)\n'))
            if search==99:
                break
            del book[search-1]
    elif check==99:
        return
    else:
        print('정확한 숫자를 입력해주세요.')
        
def show_train():
    f = open('TrainList.txt','r')
    trainlist = f.readlines()
    for i in range(1,len(trainlist)):
        trainlist[i] = trainlist[i].rstrip()
        print(trainlist[i])

def menu():
    while True:
        choose = int(input('사용할 기능을 숫자로 입력하세요.\n 1.빠른시간 기차 검색 및 예매\n 2.전체 기차 리스트 출력\n 3.나의 예매 현황 출력 및 예매 취소\n 4.프로그램 종료\n'))

        if choose == 1:
            train_search_book()
        elif choose==2:
            show_train()
        elif choose==3:
            train_book()
        elif choose==4:
            break
        else:
            print('정확한 숫자를 입력해주세요.')
try:
    menu()
except:
    menu()
    pass
```
