```python
def add_book():
    bookname, name, year, publisher, genre=input('도서명,저자,출판연도,출판사명,장르를 입력하세요.\n').split(' ')
    f = open("input1.txt",'a')
    f.writelines(' '.join([bookname,name,year,publisher,genre]))
    f.write('\n')
                 
def find_book():
    search = int(input('검색기준을 숫자로 입력하세요.\n 1.도서명\n 2.저자\n 3.출판연도\n 4.출판사명\n 5.장르\n '))  
   
    f = open('input1.txt','r')
    booklist = f.readlines()
    for i in range(0, len(booklist)) :
        booklist[i] = booklist[i].strip().split(' ') 
        
    if search == 1:
        findbookname=input('도서명을 입력하세요.\n')
        for a in range(0,len(booklist)):
            if findbookname==booklist[a][0]:
                print(booklist[a])
                
    elif search==2:
        findname=input('저자를 입력하세요.\n')
        for b in range(0,len(booklist)):
            if findname==booklist[b][1]:
                print(booklist[b])

    elif search==3:
        findyear=input('출판년도를 입력하세요.\n')
        for c in range(0,len(booklist)):
            if findyear==booklist[c][2]:
                print(booklist[c])

    elif search==4:
        findpublisher=input('출판사명을 입력하세요.\n')
        for d in range(0,len(booklist)):
            if findpublisher==booklist[d][3]:
                print(booklist[d])

    elif search==5:
        findgenre=input('장르를 입력하세요.\n')
        for f in range(0,len(booklist)):
            if findgenre==booklist[f][4]:
                print(booklist[f])

def modify_book():
    search = input('변경하고자하는 책 제목을 입력하시오.\n')
    
    f = open('input1.txt','r',encoding='utf-8')
    booklist = f.readlines()
    edited_lines = []
    cbookname, cname, cyear, cpublisher, cgenre=input('변경할 도서명,저자,출판연도,출판사명,장르를 입력하세요.\n').split(' ')
    for line in booklist:
        if search in line:
            
            edited_lines.append(cbookname)
            edited_lines.append(cname)
            edited_lines.append(cyear)
            edited_lines.append(cpublisher)
            edited_lines.append(cgenre)
            edited_lines.append('\n')
            
        else:
            edited_lines.append(line)
            
    f = open("input1.txt",'w')
    f.writelines(edited_lines)
    
def delete_book ():
    search = input('지우고자하는 책 제목을 입력하시오.\n')
    f = open('input1.txt','r',encoding='utf-8')
    booklist = f.readlines()
    edited_lines = []
    
    for line in booklist:
        if search in line:
            edited_lines.append('\0')
        else:
            edited_lines.append(line)
            
    f = open("input1.txt",'w')
    f.writelines(edited_lines)
    
def now_book():
    f = open('input1.txt','r')
    booklist = f.readlines()
    for i in range(len(booklist)):
        booklist[i] = booklist[i].rstrip()
    print(booklist)

    
def save_book():
    f.close
```
